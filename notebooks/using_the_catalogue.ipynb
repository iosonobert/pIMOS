# Using the catalogue version - 1.1.dev2 and on

```python
import pIMOS, afloat
import sys, os, glob
import pandas as pd, numpy as np
import pIMOS.utils.catalogue as pcat

project_root = r'/mnt/r/Experiments/bb23'

root_folder = os.path.join(project_root, 'data', 'moored_instrument_data')
root_folder 

os.path.exists(root_folder)
```

# Load the base catalogue
This taks an input root folder and generates a pandas dataframe from the folder structure and file naming convention. The root folder should contain a subfolder
called 'pimos_vX.Y', where X.Y is the version of the pIMOS library you are using. 


```python
cat = pcat.get_pimos_catalogue(root_folder, is_moored=True)
cat
```

# Filter the catalogue
This is just an example of the syntax for the included filter function, but it's also advisable to do a basic filter before loading the detailed data [next step].  

```python

filt_dict = {'File version': ['FV01', 'FV02'], 'instrument_model': 'Longranger'}
cat_f = pcat.filter_catalogue(cat, filt_dict)
cat_f
```

# Load the detailed metadata
Detailed metadata is that which is in the file header only, an so the file must be loaded.
 

```python
cat_f = pcat.load_detailed_data(cat_f)
cat_f
```
